```python
def find_unique_values(df, column_name):
  """
  Finds the unique values in a specified column of a Pandas DataFrame.

  Args:
    df: The input DataFrame.
    column_name: The name of the column to analyze.

  Returns:
    A list of unique values in the specified column.
    Returns an empty list if the column does not exist.
  """
  if column_name not in df.columns:
    return []
  return df[column_name].unique().tolist()

# Filter for the imporatnt columns
device_df_processed = device_df.applymap(lambda x: x.strip("b'") if isinstance(x, str) else x)
device_df_processed = device_df_processed[["Brand", "Model Name", "Models", "OS", "Type", "Internal", "SIM", "Technology", "Dimensions", "Weight"]]

# # Check Brand/Model Keywords (if you have them)
# for name in device_df_processed.columns:
#   print(name, find_unique_values(device_df_processed, name))
```

# Cleaning Dictionary Function

```python
def clean_dictionary(column: pd.Series, col_name: str) -> pd.Series:
    """
    Cleans a column in the device dictionary based on the column type.

    Args:
        column (pd.Series): The column to clean.
        col_name (str): The name of the column.

    Returns:
        pd.Series: The cleaned column.
    """
    col_name = col_name.lower().strip()

    # General cleanup for string-based columns
    if col_name in {"brand", "model name", "models", "os", "type", "sim", "technology"}:
        return column.str.lower().str.strip().fillna("unknown")

    # Handle missing values for specific string-based columns
    if col_name == "internal":
        return column.fillna("unknown").str.lower().str.strip()

    # Handle numeric or mixed fields
    if col_name in {"weight", "dimensions"}:
        return column.fillna("unknown")

    # For columns not explicitly handled
    return column
```

# Cleaning Type Column- extracting important display type information

```python
#Cleaning Type column - extracting display type for the relevant information
def standardize_type(type_value):
    """ Extract core display type from type column """
    match = re.search(r'(amoled|ips lcd|tft|lcd|monochrome|cstn|super amoled)', type_value)
    return match.group(1) if match else 'unknown'
```

# Cleaning os and standardizing it. 

```python
def standardize_os(os_name):
    """
    Standardizes OS names based on predefined rules, extracting versions and details where applicable.
    """

    if not isinstance(os_name, str):
        # Handle NaN or non-string values as "Unknown"
        return "unknown"

    os_name = os_name.lower()

    # Handle Android
    if "android" in os_name:
        match = re.search(r'android (\d+\.\d+)', os_name)
        if match:
            return f"Android {match.group(1)}"
        return "Android"

    # Handle iOS/iPadOS
    if "ios" in os_name or "ipados" in os_name:
        match = re.search(r"(ios|ipados) (\d+(\.\d+)?)", os_name)
        if match:
            return f"{match.group(1).capitalize()} {match.group(2)}"
        return "iOS"

    # Handle Windows
    if "windows" in os_name:
        if "mobile" in os_name:
            match = re.search(r"windows mobile (\d+(\.\d+)?)", os_name, re.IGNORECASE)
            version = match.group(1) if match else ""
            details = []
            if "professional" in os_name:
                details.append("Professional")
            if "standard" in os_name:
                details.append("Standard")
            if "pocketpc" in os_name:
                details.append("PocketPC")
            if "smartphone" in os_name:
                details.append("Smartphone")
            return f"Windows Mobile {version} {' '.join(details)}".strip()
        elif "phone" in os_name:
            match = re.search(r"windows phone (\d+(\.\d+)?)", os_name, re.IGNORECASE)
            version = match.group(1) if match else ""
            return f"Windows Phone {version}".strip()
        elif "rt" in os_name:
            return "Windows RT"
        elif "ce" in os_name:
            match = re.search(r"windows ce (\d+(\.\d+)?)", os_name, re.IGNORECASE)
            version = match.group(1) if match else "CE"
            return f"Windows CE {version}"
        return "Windows"

    # Handle Symbian
    if "symbian" in os_name:
        match_version = re.search(r"(symbian\s?os|symbian)\s*(\d+(\.\d+)?)?", os_name)
        match_series = re.search(r"(series\s?\d+)", os_name)
        match_uiq = re.search(r"(uiq\s?\d+(\.\d+)?)", os_name)

        # Construct the standardized name
        details = []
        if match_version:
            details.append(f"Symbian OS {match_version.group(2) or ''}".strip())
        if match_series:
            details.append(match_series.group(1).title())
        if match_uiq:
            details.append(match_uiq.group(1).upper())

        # Combine extracted details
        return " - ".join(filter(None, details)).strip()

    # Handle Blackberry
    if "blackberry" in os_name:
        match = re.search(r"blackberry os (\d+(\.\d+)?)", os_name)
        if match:
            return f"BlackBerry OS {match.group(1)}"
        return "BlackBerry OS"

    # Handle KaiOS
    if "kaios" in os_name:
        match = re.search(r"kaios (\d+(\.\d+)?)", os_name)
        if match:
            return f"KaiOS {match.group(1)}"
        return "KaiOS"

    # Handle Firefox OS
    if "firefox" in os_name:
        match = re.search(r"firefox os (\d+(\.\d+)?)", os_name)
        if match:
            return f"Firefox OS {match.group(1)}"
        return "Firefox OS"

    # Handle Tizen
    if "tizen" in os_name:
        match = re.search(r"tizen (\d+(\.\d+)?)", os_name)
        if match:
            return f"Tizen {match.group(1)}"
        return "Tizen"

    # Handle Sailfish
    if "sailfish" in os_name:
        match = re.search(r"sailfish (\d+(\.\d+)?)", os_name)
        if match:
            return f"Sailfish OS {match.group(1)}"
        return "Sailfish OS"

    # Handle Palm OS
    if "palm" in os_name:
        match = re.search(r"palm os (\d+(\.\d+)?)", os_name)
        if match:
            return f"Palm OS {match.group(1)}"
        return "Palm OS"

    # Handle MeeGo
    if "meego" in os_name:
        match = re.search(r"meego (\d+(\.\d+)?)", os_name)
        if match:
            return f"MeeGo {match.group(1)}"
        return "MeeGo"

    # Handle Nucleus OS
    if "nucleus" in os_name:
        return "Nucleus OS"

    # Handle proprietary OS
    if "proprietary os" in os_name or "proprietary" in os_name:
        return "Proprietary OS"

    # Handle Linux-based OS
    if "linux" in os_name:
        if "wearable" in os_name:
            return "Linux-based Wearable OS"
        return "Linux-based OS"

    # If no match is found, return the original string
    return os_name
```

```python
#SIM cleaning
def standardize_sim(sim_info):
    """
    Standardizes SIM information while retaining details about the type of SIM(s) and configuration.
    """

    if not isinstance(sim_info, str) or sim_info.strip() == "":
        return "unknown"

    sim_info = sim_info.lower()
    standardized = []

    # Extract SIM count
    if "quad sim" in sim_info:
        standardized.append("Quad SIM")
    elif "triple sim" in sim_info:
        standardized.append("Triple SIM")
    elif "dual sim" in sim_info:
        standardized.append("Dual SIM")
    elif "single sim" in sim_info:
        standardized.append("Single SIM")

    # Extract SIM sizes
    sim_sizes = []
    if "nano-sim" in sim_info:
        sim_sizes.append("Nano SIM")
    if "micro-sim" in sim_info:
        sim_sizes.append("Micro SIM")
    if "mini-sim" in sim_info:
        sim_sizes.append("Mini SIM")
    if "esim" in sim_info:
        sim_sizes.append("eSIM")

    if sim_sizes:
        standardized.append(" / ".join(sim_sizes))

    # Extract configuration (e.g., dual stand-by, hybrid)
    if "dual stand-by" in sim_info:
        standardized.append("Dual Stand-by")
    if "hybrid" in sim_info:
        standardized.append("Hybrid")
    if "built-in" in sim_info:
        standardized.append("Built-in")
    if "pre-installed" in sim_info:
        standardized.append("Pre-installed")

    # If no matches are found, return the original string
    return " - ".join(standardized) if standardized else sim_info
```

```python
def technology_clean(tech_value):
    if not isinstance(tech_value, str) or pd.isna(tech_value):
        return "unknown"

    tech_value = tech_value.lower()
    tech_value = tech_value.replace(" / ", "/").replace(" /", "/").replace("/ ", "/")

    tech_parts = tech_value.split("/")
    unique_tech = sorted(set(tech_parts))

    return " - ".join(unique_tech)
```

```python
def clean_dimension(value):
    """
    Cleans the dimension string to extract only the numeric parts
    and formats it as 'L x W x H mm'.
    Args:
        value (str): The raw dimension string.
    Returns:
        str: A cleaned and standardized dimension string or 'Unknown' if invalid.
    """
    if not isinstance(value, str) or value.strip() == "" or value.strip() == "-":
        return "Unknown"

    # Regex to match dimensions in 'L x W x H mm' format
    dimension_pattern = re.compile(r"(\d+(\.\d+)?)\s*x\s*(\d+(\.\d+)?)\s*x\s*(\d+(\.\d+)?)\s*mm")
    match = dimension_pattern.search(value)

    if match:
        length, _, width, _, height, _ = match.groups()
        return f"{length} x {width} x {height} mm"
    elif "thickness" in value.lower():
        # Handle cases like '5.1 mm thickness'
        thickness_match = re.search(r"(\d+(\.\d+)?)\s*mm", value)
        if thickness_match:
            thickness = thickness_match.group(1)
            return f"0 x 0 x {thickness} mm"

    return "unknown"
```

```python
def clean_weight(value):
    """
    Cleans the weight string to extract only the numeric weight in grams.
    Args:
        value (str): The raw weight string.
    Returns:
        str: A cleaned and standardized weight string in grams or 'Unknown' if invalid.
    """
    if not isinstance(value, str) or value.strip() == "" or value.strip() == "-":
        return "unknown"

    # Regex to extract weight in grams (e.g., '196 g' or '196g')
    weight_pattern = re.compile(r"(\d+(\.\d+)?)\s*g")
    match = weight_pattern.search(value)

    if match:
        weight_in_grams = match.group(1)
        return f"{weight_in_grams} g"

    return "unknown"
```

```python
def clean_weight(value):
    """
    Cleans the weight string to extract only the numeric weight in grams.

    Args:
        value (str): The raw weight string.

    Returns:
        str: A cleaned and standardized weight string in grams or 'Unknown' if invalid.
    """
    if not isinstance(value, str) or value.strip() == "" or value.strip() == "-":
        return "unknown"

    # Regex to extract weight in grams (e.g., '196 g' or '196g')
    weight_pattern = re.compile(r"(\d+(\.\d+)?)\s*g")
    match = weight_pattern.search(value)

    if match:
        weight_in_grams = match.group(1)
        return f"{weight_in_grams} g"

    return "unknown"
```

```python
def clean_model_name(model_name, brand_name):
    """
    Removes the brand name from the start of the model name if it exists.

    Parameters:
    model_name (str): The raw model name.
    brand_name (str): The brand name to remove from the model name.

    Returns:
    str: A cleaned model name with the brand name removed, if present.
    """
    if not isinstance(model_name, str) or not isinstance(brand_name, str):
        return model_name

    # Normalize both brand and model name for comparison
    model_name = model_name.lower().strip()
    brand_name = brand_name.lower().strip()

    # Check if the model name starts with the brand name
    if model_name.startswith(brand_name):
        # Remove the brand name and any leading separators (like spaces, dashes, underscores)
        cleaned_model_name = model_name[len(brand_name):].strip(" -_")
        return cleaned_model_name

    return model_name
```

```python
for column in device_df_processed.columns:
  device_df_processed[column] = clean_dictionary(device_df_processed[column], column)

device_df_processed['Model Name'] = device_df_processed.apply(
    lambda row: clean_model_name(row['Model Name'], row['Brand']),
    axis=1
)
device_df_processed['Type'] = device_df_processed['Type'].apply(standardize_type)
device_df_processed['OS'] = device_df_processed['OS'].apply(standardize_os)
device_df_processed['SIM'] = device_df_processed['SIM'].apply(standardize_sim)
device_df_processed['Technology'] = device_df_processed['Technology'].apply(technology_clean)
device_df_processed['Dimensions'] = device_df_processed['Dimensions'].apply(clean_dimension)
device_df_processed['Weight'] = device_df_processed['Weight'].apply(clean_weight)

device_df_processed.sample(30)
```
